```python

class ReLU(Module):
    """
    Rectified Linear Unit function
    """

    def __init__(self):
        self.cache = dict()
        self.grads = dict()

    def _forward(self, x):
        """
        applies relu function on x

        Args:
            x : a batch of data

        Returns:
            y : relu of input
        """
        # print()

        # print(x)
        self.cache = {
            'x': x,
        }        
        return np.maximum(x, 0)

    def backward(self, upstream):
        """
        calculate and store gradient of loss w.r.t module input

        Args:
            upstream : gradient of loss w.r.t modele output
        """
        #########################################
        ##              Your Code              ##
        #########################################
        relu_func = np.vectorize(lambda x : np.maximum(x, 0)*1/x)
        grad_x = upstream * relu_func(self.cache['x'])        
        self.grads['x'] = grad_x
```

## Train and Test Model 

### Batch Sampler


```python
import random
class RandomSampler(object):
    def __init__(self, batch_size, dataset, type):
        """
        Args:
            batch_size : sampler batch size
            dataset : dataset we want to get batch from that
            type : one of {'train', 'test', 'val'}
        """
        self.batch_size = batch_size
        self.dataset = dataset
        self.x_key = f'X_{type}'
        self.y_key = f'Y_{type}'
        self.indices = None
        self.num_batches = None

        indices = list(range(len(dataset[self.x_key])))
        self.num_batches = int(len(dataset[self.x_key])/batch_size)
        random.shuffle(indices)
        ind = []
        for i in range(self.num_batches):
            ind.append([i, indices[:batch_size]])
            indices = indices[batch_size:]
        self.indices = ind

    def __len__(self):
        assert type(self.num_batches) == int
        return self.num_batches

    def __iter__(self):
        """
        This function call when we iterate an object of this class and
        yields one batch on each call.

        Yields:
            (x, y) : tuple of bathces of x and y
        """
        
        for idx in self.indices:
            x = self.dataset[self.x_key][idx[-1]]
            y = self.dataset[self.y_key][idx[-1]]
            yield x, y

```

```python
def update_confusion_matrix(conf_matrix, preds, reals):
    """
    Updates confusion matrix

    Args:
        conf_matrix : input confusion matrix
        preds : array of predicted labels
        reals : array of real labels

    Returns:
        conf_matrix : updated confusion matrix
    """
    for i in range(len(preds)):
        pr = preds[i]
        rl = reals[i]
        # print("pr", pr, " rl", rl)
        conf_matrix[pr][rl] += 1
        # print(conf_matrix)
        # print("\n\n")
    return conf_matrix

# def calculate_acc(confusion_matrix):
def f1_score(confusion_matrix):
    """
    calculate macro f1 score from given confusion matrix

    Args:
        confusion_matrix : given confusion matrix

    Returns:
        f1 : macro f1 score
    """
    f1_score = 0
    index = 0

    for i in range(confusion_matrix.shape[0]):
        tp = confusion_matrix[i, i]
        fp = np.sum(confusion_matrix[:, i]) - tp
        fn = np.sum(confusion_matrix[i, :]) - tp
        try:
            precision = tp/(tp+fp)
        except:
            precision = 0
            print("heloooooooooo precision")
        try:
            recall = tp/(tp+fn)
        except:
            print("heloooooooooo recall")
            recall = 0
        try:
            f1_score += 2 * (precision * recall) / (precision + recall)
        except:
            f1_score += 0
    return f1_score/10

def calculate_acc(preds, reals):
    correct = 0
    total = reals.shape[0]
    # print(total)
    # print('preds', preds)
    # print('reals', reals)
    for i in range(len(preds)):
        if preds[i] == reals[i]:
            correct += 1
    
    return correct/total
```

### Define Model
Define an MLP model to solve classification problem.

```python
##          Hyper parameters           ##

n_epochs = 80
batch_size = 64
lr = 0.0021
reg_coeff = 0.22
```

```python

initializer = lambda x, y: np.random.normal(size=(y, x))

layers = [
    LinearLayer(3072, 512, initializer, True, reg_coeff),
    ReLU(),
    LinearLayer(512, 10, initializer, True, reg_coeff),
    LogSoftMax(),    

]
model = MLPModel(layers=layers)
```

```python

optimizer = Optimizer(model.get_parameters(), 'sgd', lr = lr)
criterion = CrossEntropyLoss()
train_sampler = RandomSampler(batch_size=batch_size, dataset=data, type='train')
val_sampler = RandomSampler(batch_size=batch_size, dataset=data, type='valid')
test_sampler = RandomSampler(batch_size=batch_size, dataset=data, type='test')
```

### Train Model



```python
##      Train and Validation loop      ##
train_losses, val_losses = [], []
train_accs, val_accs = [], []
train_f1, val_f1 = [], []
best_model = None
best_f1 = 0

for epoch in range(n_epochs):
    
    # Train Phase
    total_loss = 0
    N = 0
    confusion = np.zeros((10, 10))
    with tqdm.tqdm(enumerate(train_sampler), total=len(train_sampler)) as pbar:
        for i, (x, y) in pbar:
            outputs = model._forward(x)
            loss = criterion._forward(outputs, y)
            criterion.backward(1)
            model.backward(-criterion.grads['x'])
            optimizer.step()
            confusion = update_confusion_matrix(confusion, np.argmax(outputs, axis = 1), y)
            acc = calculate_acc(np.argmax(outputs, axis = 1), y)
            f1 = f1_score(confusion)
            N += 1
            total_loss += np.sum(loss)
            pbar.set_description(f'Training {epoch+1} | Loss:{total_loss/N:.2e} | Acc: {acc:.2f}| F1: {f1:.2f}|')
    
    # save epoch metrics for train phase
    train_losses.append(total_loss/N)
    train_accs.append(acc)
    train_f1.append(f1)
    

    # Validation Phase
    total_loss = 0
    N = 0
    confusion = np.zeros((10, 10))
    with tqdm.tqdm(enumerate(val_sampler), total=len(val_sampler)) as pbar:
        for i, (x, y) in pbar:
            outputs = model._forward(x)
            loss = criterion._forward(outputs, y)
            confusion = update_confusion_matrix(confusion, np.argmax(outputs, axis = 1), y)
            acc = calculate_acc(np.argmax(outputs, axis = 1), y)
            f1 = f1_score(confusion)
            N += 1
            total_loss += np.sum(loss)
            pbar.set_description(f'Validation {epoch+1} | Loss:{total_loss/N:.2e} | Acc: {acc:.2f}| F1: {f1:.2f}|')
    # save epoch metrics for validation phase
    val_losses.append(total_loss/N)
    val_accs.append(acc)
    val_f1.append(f1)
    
    
    if f1 > best_f1:
        best_f1 = f1
        best_model = copy.deepcopy(model)
    
    print(f'----------------------------[Epoch{epoch+1} finished!]----------------------------')
```
